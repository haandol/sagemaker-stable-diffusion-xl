# 모델 로드

```python
cache_dir = "/home/ec2-user/SageMaker/.cache"
```

```python
model_id = "stabilityai/stable-diffusion-xl-base-1.0"
vae_id = "madebyollin/sdxl-vae-fp16-fix"

vae = AutoencoderKL.from_pretrained(vae_id, torch_dtype=torch.float16)

pipe = DiffusionPipeline.from_pretrained(
    model_id,
    vae=vae,
    torch_dtype=torch.float16,
    variant='fp16',
    cache_dir=cache_dir,
)

pipe.load_lora_weights(
    './models',
    adapter_name="lora",
    cache_dir=cache_dir,
)
pipe.load_lora_weights(
    "latent-consistency/lcm-lora-sdxl",
    adapter_name="lcm",
    cache_dir=cache_dir,
)
pipe.load_lora_weights(
    "nerijs/pixel-art-xl",
    adapter_name="pixel",
    cache_dir=cache_dir,
)

pipe.set_adapters(["lora", "lcm", "pixel"], adapter_weights=[1.0, .8, 1.5])
pipe.scheduler = LCMScheduler.from_config(pipe.scheduler.config)
pipe.to(device="cuda")
```

# 이미지 생성

```python
prompt = "pixel, a telephone handset"
negative_prompt = "photo, realistic, hand, cord"

num_images = 3

for i in range(num_images):
    img = pipe(
        prompt=prompt,
        negative_prompt=negative_prompt,
        num_inference_steps=4,
        guidance_scale=1.5,
    ).images[0]

    img.save(f"lcm_lora_{i}.png")
```

```python
make_image_grid(
    list(map(load_image, [f"lcm_lora_{i}.png" for i in range(num_images)])),
    1,
    3,
)
```
